```python
import gymnasium as gym
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical
import numpy as np
from pyvirtualdisplay import Display
from IPython.display import Image as IPImage # For displaying GIF
import imageio

# --- Hyperparameters ---
NUM_ENVS = 5
LEARNING_RATE = 0.001
GAMMA = 0.99  # Discount factor
N_STEPS_PER_UPDATE = 5  # Collect N steps from EACH env before updating
TOTAL_TIMESTEPS = 500000
ENTROPY_COEF = 0.01 # Entropy bonus to encourage exploration
VALUE_LOSS_COEF = 0.5

# --- Check for CUDA availability ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --- Actor-Critic Network ---
class ActorCritic(nn.Module):
    def __init__(self, obs_dim, action_dim):
        super(ActorCritic, self).__init__()
        self.shared_layers = nn.Sequential(
            nn.Linear(obs_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU()
        )
        self.actor = nn.Linear(64, action_dim)  # Outputs logits for actions
        self.critic = nn.Linear(64, 1)       # Outputs state value

    def forward(self, x):
        x = self.shared_layers(x)
        action_logits = self.actor(x)
        value = self.critic(x)
        return action_logits, value

# --- Agent ---
class A2CAgent:
    def __init__(self, obs_dim, action_dim, lr=LEARNING_RATE):
        self.device = device
        self.model = ActorCritic(obs_dim, action_dim).to(device)
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)

    def get_action_and_value(self, obs_tensor, deterministic=False):
        action_logits, value = self.model(obs_tensor)
        probs = torch.softmax(action_logits, dim=-1)
        dist = Categorical(probs)

        if deterministic:
            action = torch.argmax(probs, dim=-1)
        else:
            action = dist.sample()

        log_prob = dist.log_prob(action)
        entropy = dist.entropy()
        return action, log_prob, entropy, value

    def get_value(self, obs_tensor):
        _, value = self.model(obs_tensor)
        return value

    def learn(self, log_probs, values, rewards, next_values, dones, entropies):
        # Calculate advantages
        # GAE (Generalized Advantage Estimation) is common, but for simplicity, let's use N-step returns
        # For N-step returns, we'd effectively be calculating:
        # target_q_values = rewards + GAMMA * next_values * (1 - dones)
        # advantages = target_q_values - values

        # Calculate returns (targets for value function)
        # The 'dones' tensor here should be 0 if not done, 1 if done.
        # We need to ensure rewards, next_values, dones, values are all correctly shaped.
        # next_values shape: (N_STEPS_PER_UPDATE, NUM_ENVS, 1)
        # rewards shape: (N_STEPS_PER_UPDATE, NUM_ENVS)
        # dones shape: (N_STEPS_PER_UPDATE, NUM_ENVS)

        # Let's ensure rewards and dones are correctly shaped for broadcasting
        rewards = rewards.unsqueeze(-1) # (N_STEPS_PER_UPDATE, NUM_ENVS, 1)
        dones = dones.unsqueeze(-1)     # (N_STEPS_PER_UPDATE, NUM_ENVS, 1)

        # Ensure values are also (N_STEPS_PER_UPDATE, NUM_ENVS, 1)
        # values are (N_STEPS_PER_UPDATE, NUM_ENVS, 1)
        # log_probs are (N_STEPS_PER_UPDATE, NUM_ENVS)
        # entropies are (N_STEPS_PER_UPDATE, NUM_ENVS)

        # Compute returns: R_t = r_t + gamma * V(s_{t+1}) if not done, else r_t
        # The last 'next_value' is bootstrapped from the model itself for the last step in the rollout.
        # The values in `next_values` are V(s_{t+1}), V(s_{t+2}), ..., V(s_{t+N}), last_value_bootstrapped
        returns = rewards + GAMMA * next_values * (1.0 - dones)

        # Advantages A_t = R_t - V(s_t)
        advantages = returns - values
        advantages = advantages.detach() # Stop gradients from flowing into value network through advantages

        # Policy Loss (Actor Loss)
        # We want to maximize E[A_t * log_pi(a_t|s_t)], so minimize -E[...]
        policy_loss = -(advantages * log_probs.unsqueeze(-1)).mean() # unsqueeze log_probs to match advantages shape

        # Value Loss (Critic Loss)
        value_loss = nn.functional.mse_loss(values, returns.detach()) # Detach returns for value loss

        # Entropy Loss (to encourage exploration)
        entropy_loss = -entropies.mean()

        # Total Loss
        total_loss = policy_loss + VALUE_LOSS_COEF * value_loss + ENTROPY_COEF * entropy_loss

        # Optimization
        self.optimizer.zero_grad()
        total_loss.backward()
        # Optional: Gradient clipping
        # torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=0.5)
        self.optimizer.step()

        return policy_loss.item(), value_loss.item(), entropy_loss.item()

try:
    envs = gym.make_vec(
        "CartPole-v1",
        num_envs=NUM_ENVS,
        vectorization_mode="async", # or "sync_vector_env"
    )
except Exception as e:
    print(f"Failed to create async_vector_env, trying sync_vector_env. Error: {e}")
    envs = gym.make_vec(
        "CartPole-v1",
        num_envs=NUM_ENVS,
        vectorization_mode="async"
    )


obs_space = envs.observation_space
action_space = envs.action_space

print(f"Observation space: {obs_space}, shape: {obs_space.shape}")
print(f"Action space: {action_space}, shape: {action_space.shape}, n: {action_space.nvec if hasattr(action_space, 'nvec') else action_space.n}")

# For CartPole, obs_space.shape is (NUM_ENVS, 4) and action_space is MultiDiscrete([2]*NUM_ENVS)
# We need the dimension of a single environment's observation space and action space
single_obs_dim = obs_space.shape[1]
single_action_dim = action_space.nvec[0] # For MultiDiscrete, nvec gives num_actions for each env

agent = A2CAgent(single_obs_dim, single_action_dim)

# --- Training Loop ---
# ... (initial setup) ...
obs, info = envs.reset(seed=[42 + i for i in range(NUM_ENVS)])
obs_tensor = torch.FloatTensor(obs).to(device)

# ... (tracking variables) ...

for global_step in range(0, TOTAL_TIMESTEPS, NUM_ENVS * N_STEPS_PER_UPDATE):
    # --- Collect N_STEPS_PER_UPDATE transitions from each environment ---
    log_probs_list = []
    values_list = []
    rewards_list = []
    dones_list = []
    entropies_list = []
    episode_rewards =[]
    episode_lengths =[]

    for step_idx in range(N_STEPS_PER_UPDATE): # Renamed 'step' to 'step_idx' to avoid conflict
        # Get actions, log_probs, entropy, and values WITH gradient tracking
        actions_tensor, log_prob, entropy, value = agent.get_action_and_value(obs_tensor)

        # Detach actions for environment interaction before converting to NumPy
        actions_np = actions_tensor.detach().cpu().numpy()

        # Step in the vectorized environment
        next_obs, rewards, terminations, truncations, infos = envs.step(actions_np)
        dones = np.logical_or(terminations, truncations)

        # Store experience
        log_probs_list.append(log_prob)         # Will have grad_fn
        values_list.append(value.squeeze(-1))   # Will have grad_fn
        rewards_list.append(torch.FloatTensor(rewards).to(device)) # No grad needed
        dones_list.append(torch.FloatTensor(dones).to(device))   # No grad needed
        entropies_list.append(entropy)          # Will have grad_fn

        # Update current obs
        obs = next_obs
        obs_tensor = torch.FloatTensor(obs).to(device)


    # --- Prepare data for learning ---
    log_probs_tensor = torch.stack(log_probs_list)
    values_tensor = torch.stack(values_list).unsqueeze(-1)
    rewards_tensor = torch.stack(rewards_list)
    dones_tensor = torch.stack(dones_list)
    entropies_tensor = torch.stack(entropies_list)

    # Bootstrap the value for the last next_obs
    # THIS `torch.no_grad()` IS CORRECT because last_values_bootstrapped is part of the target.
    with torch.no_grad():
        last_values_bootstrapped = agent.get_value(obs_tensor)

    next_values_list = values_list[1:]
    next_values_list.append(last_values_bootstrapped.squeeze(-1))
    next_values_tensor = torch.stack(next_values_list).unsqueeze(-1)

    # --- Perform learning update ---
    policy_loss, value_loss, entropy_loss = agent.learn(
        log_probs_tensor,
        values_tensor,
        rewards_tensor,
        next_values_tensor,
        dones_tensor,
        entropies_tensor
    )

    if (global_step / (NUM_ENVS * N_STEPS_PER_UPDATE)) % 10 == 0:
        print(f"Update {global_step // (NUM_ENVS*N_STEPS_PER_UPDATE)}: Policy Loss: {policy_loss:.4f}, Value Loss: {value_loss:.4f}, Entropy: {-entropy_loss:.4f}")

print("Training finished.")
envs.close()

```

```python
def playback_agent_manual(agent_to_test, env_name="CartPole-v1", episodes=1, output_filename="playback.gif"):
    eval_env = gym.make(env_name, render_mode="rgb_array")
    all_frames = []

    print(f"Starting manual playback for {episodes} episode(s)...")
    for episode in range(episodes):
        obs, info = eval_env.reset()
        terminated = False
        truncated = False
        total_reward = 0
        episode_frames = []

        while not (terminated or truncated):
            # Render the current state
            frame = eval_env.render()
            if frame is not None:
                 episode_frames.append(frame)

            obs_tensor = torch.FloatTensor(obs).unsqueeze(0).to(agent_to_test.device)
            with torch.no_grad():
                action_tensor, _, _, _ = agent_to_test.get_action_and_value(obs_tensor, deterministic=True)
            action = action_tensor.cpu().item()

            obs, reward, terminated, truncated, info = eval_env.step(action)
            total_reward += reward

        all_frames.extend(episode_frames) # Collect frames from this episode
        print(f"Episode {episode + 1}: Total Reward: {total_reward}, Frames collected: {len(episode_frames)}")

    eval_env.close()
    print("Manual playback finished.")

    if all_frames:
        print(f"Saving {len(all_frames)} frames to {output_filename}...")
        imageio.mimsave(output_filename, all_frames, fps=30) # Adjust FPS as needed
        print(f"Saved to {output_filename}")
        return output_filename
    else:
        print("No frames collected.")
        return None
```
